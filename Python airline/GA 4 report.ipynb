```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
Ga_df = pd.read_csv(r'/Users/daviddangol/Documents/monthly report/domestic airlines/data/GA4 report .csv')
```

```python
# User ID frequency with date wise and %
user_id_date_freq = Ga_df.groupby(['event_date', 'user_id']).size().reset_index(name='frequency')
user_id_date_freq['percentage'] = user_id_date_freq['frequency'] / user_id_date_freq.groupby('event_date')['frequency'].transform('sum') * 100
display(user_id_date_freq.head().nlargest(10, 'frequency'))
```

```python
# User ID frequency count and %
user_id_freq = Ga_df['user_id'].value_counts().reset_index(name='count')
user_id_freq.columns = ['user_id', 'count']
user_id_freq['percentage'] = user_id_freq['count'] / len(Ga_df) * 100
display(user_id_freq.head().round(2))
```

```python
# User ID frequency count who have name in account_id as they made payment and %
paid_users = Ga_df[Ga_df['Account_ID'].notna()]
paid_user_freq = paid_users['user_id'].value_counts().reset_index(name='count')
paid_user_freq.columns = ['user_id', 'count']
paid_user_freq['percentage'] = paid_user_freq['count'] / len(paid_users) * 100
display(paid_user_freq.head().round(4))
```

```python
# Event date to when the next payment after event
event_payment_diff = Ga_df[['event_date', 'next_payment_after_event']].dropna()
event_payment_diff['event_date'] = pd.to_datetime(event_payment_diff['event_date'])
event_payment_diff['next_payment_after_event'] = pd.to_datetime(event_payment_diff['next_payment_after_event'])
event_payment_diff['days_to_next_payment'] = (event_payment_diff['next_payment_after_event'] - event_payment_diff['event_date']).dt.days
display(event_payment_diff.head())
```

```python
# Count of destination city and %
destination_city_freq = Ga_df['destinationCity'].value_counts().reset_index(name='count')
destination_city_freq.columns = ['destinationCity', 'count']
destination_city_freq['percentage'] = destination_city_freq['count'] / len(Ga_df) * 100
display(destination_city_freq.head().round(2))
```

```python
# Departure city and % and count
departure_city_freq = Ga_df['departureCity'].value_counts().reset_index(name='count')
departure_city_freq.columns = ['departureCity', 'count']
departure_city_freq['percentage'] = departure_city_freq['count'] / len(Ga_df) * 100
display(departure_city_freq.head().round(2))
```

```python
# User ID count to event name
user_id_event_freq = Ga_df.groupby('event_name')['user_id'].nunique().reset_index(name='unique_user_count')
display(user_id_event_freq.head())
```

```python
# Analysis with user id and event date to time to payment date
user_event_payment_time_diff = Ga_df[['user_id', 'event_date', 'payment_date']].dropna()
user_event_payment_time_diff['event_date'] = pd.to_datetime(user_event_payment_time_diff['event_date'])
user_event_payment_time_diff['payment_date'] = pd.to_datetime(user_event_payment_time_diff['payment_date'], errors='coerce')
user_event_payment_time_diff['days_to_payment'] = (user_event_payment_time_diff['payment_date'] - user_event_payment_time_diff['event_date']).dt.days
display(user_event_payment_time_diff.sort_values(by='days_to_payment', ascending=False).head())
```
