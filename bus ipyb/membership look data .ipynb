```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
bus_look= pd.read_csv(r'/Users/daviddangol/Documents/monthly report/Bus Book Data/data/bus_look_data_full.csv')
```

```python
display(bus_look.head())
display(bus_look.info())
display(bus_look.describe())
```

```python

bus_look['membership_start_npt'] = pd.to_datetime(bus_look['membership_start_npt'])
bus_look_filtered = bus_look[bus_look['membership_start_npt'].dt.date > pd.to_datetime('2025-10-15').date()]
display(bus_look_filtered.head())
```

```python
bus_look_filtered['membership_start_npt'] = pd.to_datetime(bus_look_filtered['membership_start_npt'])
bus_look_date_filtered = bus_look_filtered[(bus_look_filtered['membership_start_npt'] >= '2025-10-15') & (bus_look_filtered['membership_start_npt'] < '2025-11-01')]
display(bus_look_date_filtered.head())
display(bus_look_date_filtered.info())
```

```python
unique_user_count = bus_look_date_filtered['user_id'].nunique()
print(f"The number of unique users in the filtered data is: {unique_user_count}")

given_count = 17000
if abs(unique_user_count - given_count) < given_count * 0.1:
    print(f"The calculated unique user count ({unique_user_count}) is close to the given value ({given_count}).")
else:
    print(f"The calculated unique user count ({unique_user_count}) is significantly different from the given value ({given_count}).")
```

```python
# 1. Analyze the distribution of operating systems
os_counts = bus_look_date_filtered['operating_system'].value_counts()
os_percentages = bus_look_date_filtered['operating_system'].value_counts(normalize=True) * 100
print("Operating System Distribution:")
display(os_counts)
display(os_percentages)
```

```python

# 2. Analyze the distribution of countries
country_counts = bus_look_date_filtered['country'].value_counts()
country_percentages = bus_look_date_filtered['country'].value_counts(normalize=True) * 100
print("\nTop 10 Countries Distribution:")
display(country_counts.head(10))
display(country_percentages.head(10))
```

```python
# 3. Analyze the distribution of cities
city_counts = bus_look_date_filtered['city'].value_counts()
city_percentages = bus_look_date_filtered['city'].value_counts(normalize=True) * 100
print("\nTop 10 Cities Distribution:")
display(city_counts.head(10))
display(city_percentages.head(10))
```

```python

# 4. Analyze the distribution of the day of the week for membership_start_npt
bus_look_date_filtered['start_dayofweek'] = bus_look_date_filtered['membership_start_npt'].dt.day_name()
dayofweek_counts = bus_look_date_filtered['start_dayofweek'].value_counts()
dayofweek_percentages = bus_look_date_filtered['start_dayofweek'].value_counts(normalize=True) * 100
print("\nMembership Start Day of the Week Distribution:")
display(dayofweek_counts)
display(dayofweek_percentages)
```

```python

# 5. Analyze the distribution of the hour of the day for membership_start_npt
bus_look_date_filtered['start_hour'] = bus_look_date_filtered['membership_start_npt'].dt.hour
hour_counts = bus_look_date_filtered['start_hour'].value_counts().sort_index()
hour_percentages = bus_look_date_filtered['start_hour'].value_counts(normalize=True).sort_index() * 100
print("\nMembership Start Hour of the Day Distribution:")
display(hour_counts)
display(hour_percentages)
```
