```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
qfx = pd.read_csv(r'/Users/daviddangol/Documents/monthly report/Qfx/data/movies analysis sept to oct .csv')
```

```python
qfx['txn_date'] = pd.to_datetime(qfx['txn_date'], format='%d/%m/%Y')
qfx['day_of_week'] = qfx['txn_date'].dt.day_name()
```

```python
min_date = qfx['txn_date'].min()
max_date = qfx['txn_date'].max()

print(f"Minimum Transaction Date (QFX Clients): {min_date}")
print(f"Maximum Transaction Date (QFX Clients): {max_date}")
```

```python
before_gprs = qfx[qfx['txn_date'] <= '2025-09-16']
after_gprs = qfx[qfx['txn_date'] > '2025-09-17']

print("Dataframe 'before_gprs' (transactions up to 2025-09-16):")
display(before_gprs.head())

print("\nDataframe 'after_gprs' (transactions after 2025-09-17):")
display(after_gprs.head())
```

```python
before_gprs['channel'].value_counts()
```

```python
after_gprs['channel'].value_counts()
```

```python
theater_counts = after_gprs['theater'].value_counts()
theater_percentages = after_gprs['theater'].value_counts(normalize=True) * 100

theater_summary = pd.DataFrame({
    'Count': theater_counts,
    'Percentage': theater_percentages.round(2) 
})

print("Theater Transaction Summary (after 2025-09-23):")
display(theater_summary.head(10))
```

```python
gprs_data = after_gprs[after_gprs['channel'] == 'GPRS']
display(gprs_data.head())
```

```python
# Get unique account IDs from both dataframes
before_gprs_accounts = set(before_gprs['account_id'].unique())
gprs_data_accounts = set(gprs_data['account_id'].unique())

# Find common account IDs
common_accounts = list(before_gprs_accounts.intersection(gprs_data_accounts))

print(f"Number of common account IDs: {len(common_accounts)}")

# Calculate frequency of common accounts in both dataframes
before_gprs_common_frequency = before_gprs[before_gprs['account_id'].isin(common_accounts)]['account_id'].value_counts()
gprs_data_common_frequency = gprs_data[gprs_data['account_id'].isin(common_accounts)]['account_id'].value_counts()

print("\nFrequency of common account IDs in 'before_gprs':")
display(before_gprs_common_frequency.head())

print("\nFrequency of common account IDs in 'gprs_data':")
display(gprs_data_common_frequency.head())
```

```python
# Get the total number of unique account IDs in gprs_data
total_gprs_unique_accounts = gprs_data['account_id'].nunique()

print(f"Total unique account IDs in 'gprs_data': {total_gprs_unique_accounts}")

# Number of common account IDs was already calculated in a previous cell
# len(common_accounts)

# Calculate the percentage of common accounts relative to total unique GPRS accounts
percentage_common_in_gprs = (len(common_accounts) / total_gprs_unique_accounts) * 100

print(f"\nPercentage of common account IDs in 'gprs_data': {percentage_common_in_gprs:.2f}%")

# You can also calculate the percentage of common accounts relative to total unique accounts before GPRS
total_before_gprs_unique_accounts = before_gprs['account_id'].nunique()
percentage_common_in_before = (len(common_accounts) / total_before_gprs_unique_accounts) * 100

print(f"Percentage of common account IDs in 'before_gprs': {percentage_common_in_before:.2f}%")

```

```python
# Filter before_gprs for 'WEB' channel transactions before 2025-09-17
before_sept_web = before_gprs[(before_gprs['txn_date'] < '2025-09-17') & (before_gprs['channel'] == 'WEB')]

# Get unique account IDs for both filtered dataframes
unique_before_sept_web_accounts = set(before_sept_web['account_id'].unique())
unique_gprs_accounts = set(gprs_data['account_id'].unique())

# Count the unique IDs
count_before_sept_web = len(unique_before_sept_web_accounts)
count_gprs = len(unique_gprs_accounts)

print(f"Number of unique account IDs in 'before_gprs' (WEB channel before 2025-09-17): {count_before_sept_web}")
print(f"Number of unique account IDs in 'gprs_data': {count_gprs}")

# Find common accounts between these two sets
common_accounts_sept_gprs_web = list(unique_before_sept_web_accounts.intersection(unique_gprs_accounts))

print(f"\nNumber of common account IDs between 'before_gprs' (WEB before 2025-09-17) and 'gprs_data': {len(common_accounts_sept_gprs_web)}")


# Filter the original dataframe to include data for the common accounts
common_sept_gprs_web_data = qfx[qfx['account_id'].isin(common_accounts_sept_gprs_web)]

# Save the filtered data to a CSV file
common_sept_gprs_web_data.to_csv('common_sept_gprs_web_transactions.csv', index=False)

print("\nData for common account IDs (before Sept 17th WEB and GPRS) saved to 'common_sept_gprs_web_transactions.csv'")
```
