```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
qfx_old = pd.read_csv('/Users/daviddangol/Downloads/movies_reconciliation_ashwin.csv')
```

```python
display(qfx_old.head())
display(qfx_old.info())
```

```python
min_date = qfx_old['Date'].min()
max_date = qfx_old['Date'].max()
```

```python
print(f"Minimum Transaction Date (QFX Clients): {min_date}")
print(f"Maximum Transaction Date (QFX Clients): {max_date}")
```

```python
# Separate analysis for 'Product Code'
product_code_analysis = qfx_old.groupby('Product Code').size().reset_index(name='Count')
product_code_analysis['Frequency'] = product_code_analysis['Count']
total_product_codes = product_code_analysis['Count'].sum()
product_code_analysis['Percentage'] = (product_code_analysis['Count'] / total_product_codes) * 100

# Sort by percentage in descending order for product code analysis
product_code_analysis = product_code_analysis.sort_values(by='Percentage', ascending=False)

print("\nAnalysis by Product Code:")
display(product_code_analysis.round(2))
```

```python
product_profile_analysis = qfx_old.groupby(['Product Code', 'Profile']).size().reset_index(name='Count')

# Filter for 'User' profile and sort by count in descending order
user_profile_analysis = product_profile_analysis[product_profile_analysis['Profile'] == 'User'].sort_values(by='Count', ascending=False)

print("Analysis for User Profiles:")
display(user_profile_analysis)
```

```python
# Filter for 'Agent' profile and sort by count in descending order
agent_profile_analysis = product_profile_analysis[product_profile_analysis['Profile'] == 'Agent'].sort_values(by='Count', ascending=False)

print("\nAnalysis for Agent Profiles:")
display(agent_profile_analysis)
```
